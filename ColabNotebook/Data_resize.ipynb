**Import Module**

```python
import os
import numpy as np
import cv2
from glob import glob
from tqdm import tqdm
import imageio
```

```python
def create_dir(path):
    if not os.path.exists(path):
        os.makedirs(path)
```

```python
def load_data(path):
    X = sorted(glob(os.path.join(path, "data/train/", "*.jpeg")))
    Y = sorted(glob(os.path.join(path, "data/train_gt/", "*.jpeg")))

    train_size = int(len(X) * 0.8)
    train_x = X[:train_size]
    train_y = Y[:train_size]

    test_x = X[train_size:]
    test_y = Y[train_size:]

    return train_x, train_y, test_x, test_y
```

```python
def create_new_data(images, masks, save_path, augment = True):
    size = (512, 512)
    for idx, (x, y) in tqdm(enumerate(zip(images, masks)), total = len(images)):
        """ Extracting the name """ 
        name = x.split("/")[-1].split(".")[0]
        # print(name)
        # break
        """ Reading image and mask """

        x = cv2.imread(x, cv2.IMREAD_COLOR)
        y = cv2.imread(y, cv2.IMREAD_GRAYSCALE)
        
        X = [x]
        Y = [y]

        index = 0
        for i, m in zip(X, Y):
            i = cv2.resize(i, size)
            m = cv2.resize(m, size)

            tmp_image_name = f"{name}_{index}.png"
            tmp_mask_name = f"{name}_{index}.png"

            image_path = os.path.join(save_path, "image", tmp_image_name)
            mask_path = os.path.join(save_path, "mask", tmp_mask_name)
# 
            cv2.imwrite(image_path, i)
            cv2.imwrite(mask_path, m)
```

```python
if __name__ == "__main__":
    path = "drive/MyDrive/BKAI_CV/"
    train_x, train_y, test_x, test_y = load_data(path)
    print(f"Train: {len(train_x)} - {len(train_y)}")
    print(f"Test: {len(test_x)} - {len(test_y)}")

    create_dir("drive/MyDrive/BKAI_CV/data_resize/train/image/")
    create_dir("drive/MyDrive/BKAI_CV/data_resize/train/mask/")
    create_dir("drive/MyDrive/BKAI_CV/data_resize/test/image/")
    create_dir("drive/MyDrive/BKAI_CV/data_resize/test/mask/")

    create_new_data(train_x, train_y,"drive/MyDrive/BKAI_CV/data_resize/train/" )
    create_new_data(test_x, test_y,"drive/MyDrive/BKAI_CV/data_resize/test/" )
```
